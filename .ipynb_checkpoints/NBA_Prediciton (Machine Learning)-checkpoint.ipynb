```python
#identifying error metric (where 1.0 is perfect)
error_metric=0
def find_ap(combination):
    actual = combination.sort_values("Share", ascending=False).head(5)
    predicted = combination.sort_values("predictions", ascending=False)
    ps = []
    found = 0
    seen = 1
    for index, row in predicted.iterrows():
        if row["Player"] in actual["Player"].values:
            found +=1 
            ps.append(found/seen)
        seen += 1
    return sum(ps) / len(ps)
```

```python
find_ap(combination)
```

```python
# implementing backtesting to predict each year
years = list(range(1991,2022))
```

```python
aps = []
all_predictions = []
for year in years[5:]: #starts with the fifth year(1996) of our data
    train = stats[stats["Year"] < year]
    test = stats[stats["Year"] == year]
    # Above shows that we train data from year (1991 - 1995) then we test data from year 1996
    # Train the model
    reg.fit(train[predictors], train["Share"])
    predictions = reg.predict(test[predictors])
    predictions = pd.DataFrame(predictions, columns=["predictions"], index=test.index)
    combination = pd.concat([test[["Player", "Share"]], predictions], axis=1)
    all_predictions.append(combination)
    aps.append(find_ap(combination))
```

```python
sum(aps) / len(aps)
# below is the mean average precision
```

```python
#diagnose the algo performance more
def add_ranks(combination):
    combination = combination.sort_values("Share", ascending=False) #sort using "Share" 
    combination["Rk"] = list(range(1,combination.shape[0]+1)) # make the rank (rk) column
    combination = combination.sort_values("predictions", ascending=False)
    combination["Predicted_Rk"] = list(range(1,combination.shape[0]+1))
    combination["Diff"] = combination["Rk"] - combination["Predicted_Rk"] # make a new column ["Diff"]
    return combination
```

```python
ranking = add_ranks(all_predictions[1])
ranking[ranking["Rk"] < 6].sort_values("Diff", ascending=False) # only take the top 5
```

```python
# build a backtesting function

def backtest(stats, model, year, predictors):
    aps = []
    all_predictions = []
    for year in years[5:]: #starts with the fifth year(1996) of our data
        train = stats[stats["Year"] < year]
        test = stats[stats["Year"] == year]
        # Above shows that we train data from year (1991 - 1995) then we test data from year 1996
        # Train the model
        model.fit(train[predictors], train["Share"])
        predictions = reg.predict(test[predictors])
        predictions = pd.DataFrame(predictions, columns=["predictions"], index=test.index)
        combination = pd.concat([test[["Player", "Share"]], predictions], axis=1)
        combination = add_ranks(combination)
        all_predictions.append(combination)
        aps.append(find_ap(combination))
    return sum(aps) / len(aps), aps, pd.concat(all_predictions)
```

```python
mean_ap, aps, all_predictions = backtest(stats, reg, years[5:], predictors)
```
